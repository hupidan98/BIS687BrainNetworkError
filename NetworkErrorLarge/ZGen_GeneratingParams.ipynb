```python
import numpy as np
from numpy.random import default_rng
import random
import RunningFFnetwork
import ConnectionMapGen

import pickle

rng = default_rng()
def poisson_process(start: float, max_t: float, avg_interval: float) -> list:
    t = start
    events = []
    while t < max_t:
        t += rng.exponential(avg_interval)
        events.append(t)
    # discard the last event, as it exceeds max_t
    if events:
        events = events[:-1]
    return events
```

```python
## Single Firing?
print('See if single fire is going to light up the next target')
########### Params
## feed forward tuneing
print('FeedForward Connect')
std_n = 38
n = 38
nlevel = 2
node_level_num = []
for i in range(nlevel):
    node_level_num.append(n)

stim_w = 0.000177 * 1

network_connmat= np.zeros((n * nlevel, n * nlevel))
network_connmat[n: , n:] = stim_w

CellType = "point"

total_ruuning_time = 10000

MeanDelay = 3
stdDelay = 0.0

stimulations = []
average_interval = 3
settle_time = 20
for i in range(node_level_num[0]):
    stimulations.append(poisson_process(settle_time, total_ruuning_time, average_interval))
    
output = RunningFFnetwork.runingFFnetwork(network_connmat, CellType, total_ruuning_time,
                    stimulations, MeanDelay, stdDelay)

for i in range(nlevel):
    showNval = 30
    for j in range(showNval):
        print(len(output[i * n + j * int(n/showNval)]), end = ' ')
    print()
```

```python
# ScaleFree
# param_list = []

for n in np.arange(200, 201, 10):
    total_ruuning_time = 10000
    average_interval = 3
    settle_time = 10
    nlevel = 1
    CellType = "point"
    node_level_num = []
    for ii in range(nlevel):
        node_level_num.append(n)
    
    stimulations_lst = []
    for input_idx in range(10):
        stimulations = []
        for idx in range(n):
            stimulations.append(poisson_process(settle_time, total_ruuning_time, average_interval))
        stimulations_lst.append(stimulations)
    
    model_conn_lst = []
    for model_id in np.arange(0, 10, 1):
        stim_w = 0.000177 * 1
        network_connmat= ConnectionMapGen.ScaleFreeMapGen(sum(node_level_num),
                          max_weight = stim_w *np.sqrt(38 / n),
                         deterministic_weight = None)
        model_conn_lst.append(network_connmat)
            
            
    for input_idx in range(10):
        stimulations = stimulations_lst[input_idx]
        for model_id in np.arange(0, 10, 1):
            network_connmat = model_conn_lst[model_id]
          
            param_list = []
            for MeanDelay_noround in np.arange(3, 3.01, 0.2):
                MeanDelay = np.round(MeanDelay_noround,1)
                for stdDelay_noround in np.arange(0, 1.01, 0.05):
                    stdDelay = np.round(stdDelay_noround,2)
                    for n_run in np.arange(0, 10, 1):
                        para = {}
                        para['CellType'] = CellType
                        para['network_connmat'] = network_connmat
                        para['total_ruuning_time'] = total_ruuning_time
                        para['stimulations'] = stimulations
                        para['MeanDelay'] = MeanDelay
                        para['stdDelay'] = stdDelay
                        para['model_idx'] = model_id
                        para['n_run'] = n_run
                        para['nlevel'] = nlevel
                        para['layercount'] = n
                        para['unique_id'] = "ScaleFree" + '_' + str(CellType) + '_layercount' + str(n) + '_model' + str(model_id) + '_input' + str(input_idx) + '_stddelay' + str(stdDelay) + '_meandelay' + str(MeanDelay) + '_nrun' + str(n_run)
                        param_list.append(para)
            print(model_id, input_idx, n, len(param_list))
            with open( './params/'+"ScaleFree" + '_' + str(CellType) + '_layercount' + str(n) +  '_model' + str(model_id) + '_input' + str(input_idx) + '_' + 'running_params.pkl', 'wb') as outp:
                pickle.dump(param_list, outp, pickle.HIGHEST_PROTOCOL) 
```
