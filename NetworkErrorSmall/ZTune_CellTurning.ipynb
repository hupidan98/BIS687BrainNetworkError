```python
from neuron import h, gui
from neuron.units import ms, mV
import Cell
import ConnectionMapGen as cmg
import spikedelayer as sd

h.load_file("stdrun.hoc")
import numpy as np
import random
```

```python
cell = Cell.PointCell(0, 0, 0, 0, 0)
cell.pos_stim.number = 1
cell.pos_stim.start = 20
cell.pos_stim.interval = 5
shape_window = h.PlotShape(True)
shape_window.show(0)

t = h.Vector().record(h._ref_t)
h.finitialize(-65 * mV)
h.continuerun(100)

%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(t, cell.soma_v)
plt.show()
```
